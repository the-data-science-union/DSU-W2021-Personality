# Personality Correlation EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
young_people_data = pd.read_csv('../cleaned_youngpeople_data.csv')
```

```python
pd.set_option('display.max_columns', 500)
```

```python
young_people_data.head()
```

```python
print(young_people_data.columns)
```

```python
sns.countplot(x="Appearence and gestures", hue="Happiness in life", data=young_people_data)
plt.title('Weak Positive Correlation between Appearance and Happiness in Life', y=1.08, fontsize=15)
plt.xlabel('"I am well-mannered and care about my appearence" (1-5)')
plt.ylabel('# of Respondants')
plt.legend(title='"I am happy in life" (1-5)')
#red line
plt.plot([0.1, 1.25], [8, 50], color='red', linestyle='--', linewidth=1)
plt.plot([1.25, 2.1], [50, 172], color='red', linestyle='--', linewidth=1)
plt.plot([2.1, 3.2], [172, 195], color='red', linestyle='--', linewidth=1)
plt.plot([3.2, 4.2], [195, 77], color='red', linestyle='--', linewidth=1)
#purple line
plt.plot([0.3, 1.35], [6, 14], color='purple', linestyle='--', linewidth=1)
plt.plot([1.35, 2.3], [14, 42], color='purple', linestyle='--', linewidth=1)
plt.plot([2.3, 3.3], [42, 50], color='purple', linestyle='--', linewidth=1)
plt.plot([3.3, 4.35], [50, 47], color='purple', linestyle='--', linewidth=1)
#plt.savefig('appearance-happiness.png')
plt.show()
```

```python
sns.countplot(x="God", hue="Final judgement", data=young_people_data)
plt.title('Moderate Positive Correlation between Faith and Final Judgement', y=1.08, fontsize=15)
plt.xlabel('"I believe in God" (1-5)')
plt.ylabel('# of Respondants')
plt.legend(title='"Bad will suffer \nand good \nwill be \nrewarded"\n(1-5)', bbox_to_anchor = [1, 1])
#blue line
plt.plot([-0.3, 0.75], [123, 55], color='teal', linestyle='--', linewidth=1)
plt.plot([0.75, 1.6], [55, 65], color='teal', linestyle='--', linewidth=1)
plt.plot([1.6, 2.7], [65, 35], color='teal', linestyle='--', linewidth=1)
plt.plot([2.7, 3.75], [35, 45], color='teal', linestyle='--', linewidth=1)
#purple line
plt.plot([0.3, 1.25], [3, 10], color='purple', linestyle='--', linewidth=1)
plt.plot([1.25, 2.25], [10, 20], color='purple', linestyle='--', linewidth=1)
plt.plot([2.25, 3.35], [20, 25], color='purple', linestyle='--', linewidth=1)
plt.plot([3.35, 4.3], [25, 85], color='purple', linestyle='--', linewidth=1)
#plt.savefig('faith.png', bbox_inches='tight')
plt.show()
```

```python
health_heatmap = young_people_data[['Spending on healthy eating', 'Healthy eating', 'Health']].pivot_table(
    values='Spending on healthy eating',
    index='Healthy eating',columns='Health',
    aggfunc=np.mean)
health_heatmap.sort_index(level=0, ascending=False, inplace=True)
plt.title('Strong Positive Correlation between Health Views and Actions', y=1.5,fontsize=15)
sns.heatmap(health_heatmap,annot=True, 
            cbar_kws = dict(use_gridspec=False,location="top",label='"I am willing to spend extra \non healthy food" (1-5)'), 
            cmap="Greens")
plt.xlabel('"I worry about my health" (1-5)')
plt.ylabel('"I eat healthy" (1-5)')
plt.ylabel('"I eat healthy" (1-5)')
#plt.savefig('health.png', bbox_inches='tight')
plt.show()
```

```python
young_people_heatmap = young_people_data[['Self-criticism', 'Personality', 'Happiness in life']].pivot_table(
    values='Self-criticism',
    index='Personality',columns='Happiness in life',
    aggfunc=np.mean)
young_people_heatmap.sort_index(level=0, ascending=False, inplace=True)
plt.title('Moderate Negative Correlation between Self-Esteem and Self-Doubt', y=1.5,fontsize=15)
sns.heatmap(young_people_heatmap,annot=True, 
            cbar_kws = dict(use_gridspec=False,location="top",label='"I often think about and regret\n the decisions I make" (1-5)'),
           cmap="Reds")
plt.xlabel('"I am happy in life" (1-5)')
plt.ylabel('"I believe all my personality traits are positive" (1-5)')
#plt.savefig('esteem-happiness.png', bbox_inches='tight')
plt.show()
```

```python
young_people_heatmap = young_people_data[['Empathy', 'Finances', 'Charity']].pivot_table(
    values='Charity',
    index='Empathy',columns='Finances',
    aggfunc=np.mean)
young_people_heatmap.sort_index(level=0, ascending=False, inplace=True)
plt.title('No Correlation between Empathy, Frugality, and Charity', y=1.45,fontsize=15)
sns.heatmap(young_people_heatmap,annot=True, cbar_kws = dict(use_gridspec=False,location="top",label='"I always give to charity" (1-5)'),
           cmap="Blues")
plt.xlabel('"I save all the money I can" (1-5)')
plt.ylabel('"I am empathetic" (1-5)')
#plt.savefig('empathy-spending-charity.png', bbox_inches='tight')
plt.show()
```
